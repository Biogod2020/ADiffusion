```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- Configuration ---
csv_file = '/Users/jay/Downloads/scalars.csv'  # Make sure this file is in the same directory as the script,
                          # or provide the full path.
output_image_file = 'adiffusion_loss_curve.png' # Name for the saved plot image
plot_title = 'ADiffusion Training Loss (on HEST-1k Dataset)'
x_label = 'Training Epoch'
y_label = 'Loss Value'
# --- End Configuration ---

# Check if the CSV file exists
if not os.path.exists(csv_file):
    print(f"Error: CSV file '{csv_file}' not found in the current directory.")
    print("Please make sure the file is present or provide the full path.")
else:
    try:
        # Load the data using pandas
        # Assumes columns are named 'Step' and 'Value' based on the CSV structure
        loss_data = pd.read_csv(csv_file)

        # Check if required columns exist
        if 'Step' not in loss_data.columns or 'Value' not in loss_data.columns:
             print(f"Error: CSV file '{csv_file}' must contain 'Step' and 'Value' columns.")
        else:
            # Create the plot
            # figsize=(7, 4) might be a good starting point for PPT slides
            fig, ax = plt.subplots(figsize=(6, 3), dpi=300)

            # Plot the loss Value against the Step
            # Added markers because there are few data points
            ax.plot(loss_data['Step'], loss_data['Value'], marker='o', linestyle='-', linewidth=2, markersize=5, color='tab:blue')

            # Customize the plot for clarity and presentation
            ax.set_title(plot_title, fontsize=14, fontweight='bold')
            ax.set_xlabel(x_label, fontsize=12)
            ax.set_ylabel(y_label, fontsize=12)
            ax.grid(True, linestyle='--', alpha=0.7) # Add a grid for better readability

            # Optional: Adjust tick label sizes if needed
            ax.tick_params(axis='both', which='major', labelsize=10)

            # Ensure layout is tight to prevent labels overlapping when saving
            plt.tight_layout()

            # Save the plot as an image file
            try:
                # dpi=300 for good resolution in presentations
                # bbox_inches='tight' helps prevent labels from being cut off
                plt.show()
                print(f"Success! Loss curve plot saved as '{output_image_file}'")
                print("You can now insert this image into your PowerPoint slide for 'ADiffusion'.")
            except Exception as e:
                print(f"Error saving plot: {e}")

            # Display the plot (optional, can be commented out if you only need the file)
            # plt.show()

    except Exception as e:
        print(f"An error occurred processing the CSV file: {e}")
```
